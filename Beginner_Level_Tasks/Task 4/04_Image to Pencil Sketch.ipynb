#### Importing libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

#### Look of image

```python
imgz = cv2.imread("Tom&Jerry.jpg")
img = imgz[:,:,::-1]
plt.imshow(img)
plt.show()
```

#### Conversion to GrayScale

```python
gray_image = cv2.cvtColor(imgz, cv2.COLOR_BGR2GRAY)
img_gray = gray_image[:,:]
plt.imshow(gray_image,cmap = 'gray')
plt.show()
```

#### Inverted Image

```python
inv_img = 255 - gray_image
plt.imshow(inv_img,cmap = 'gray')
plt.show()
```

#### Blur Image

```python
blur = cv2.GaussianBlur(inv_img,(21,21),0)
```

#### Invert The Blur Image

```python
inv_blur = 255-blur
p_sketch = cv2.divide(gray_image,inv_blur,scale = 256.0)
plt.imshow(p_sketch,cmap = 'gray')
plt.show()
```
